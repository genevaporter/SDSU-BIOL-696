(2) Transform variables (either sqrt() or log()) if non-normal and repeat histogram, qqnorm, qqline on transformed data

```r
# Transforming data sets to better fit Q-Q line 

hist(log(strep))
hist(sqrt(prev))
hist(sqrt(lepto))
hist(fuso)
hist(sqrt(veil))
hist(deepest)

qqnorm(log(strep), main="Q-Q Plot: strep (log transform)")
qqline(log(strep))

qqnorm(sqrt(prev), main="Q-Q Plot: prev (square root transform)") 
qqline(sqrt(prev))

qqnorm(sqrt(lepto), main="Q-Q Plot: lepto (square root transform)")
qqline(sqrt(lepto))

qqnorm(sqrt(veil), main="Q-Q Plot: veil (square root transform)") 
qqline(sqrt(veil))
```

## Remarks on normal distribution:

### Histograms

The Histograms show us the frequency of each species measurement for each sample taken. For veil, we can see that measurements between 5 and 10 occur 20 times in the data set (before transformation)

### qqnorm() and qqline()

This analysis tells us if our data is normally distributed, that is, it follows the bell curve model without skew. Normally distributed datasets visualized using qqnorm() will have points that are very close to the line produced by qqline(). As we can see, some visualizations did not appear linear (strep, prev, lepto, and veil). Transformation is needed to perform further analysis that requires normal distribution in the data.

### Variable transformation

The non-normal data can be transformed by taking the log or square root of the data (among other methods). For strep (skewed right), the log transform was the more appropriate representation for a better data fit. Prev, lepto, and veil (all also skewed right) were better suited to square root transformations. Note that there was a ***mostly*** good fit for a log transform of veil, but it produced a few huge outliers. Fuso did not benefit from either a log or a square root transformation, even though it was slightly skewed right. Deepest was already normalized, and was largely unaffected by either transformation.

## One-way ANOVA and summary and boxplots

```r
# Lepto by time

oneway.test(sqrt(lepto) ~  time)
summary(sqrt(lepto))
boxplot(sqrt(lepto) ~ time, main="Lepto by Time", 
        xlab = "time", ylab = "lepto")
```

```r
# Strep by time

oneway.test(log(strep) ~ time)
summary(log(strep))
boxplot(log(strep) ~ time, main="Strep by Time", 
        xlab = "time", ylab = "strep")
```

## Remarks on ANOVA:

ANOVA is the analysis of variance. Using one-way ANOVA determines if populations in a data set have different means. Here we compared populations in the first and second samples (time 1 and time 2) across the samples taken from all sites. This lets us know if a) the data collection procedures were consistent, or b) if taking a single sample from a site affected the next sample taken at the same site. (This is mostly conjecture for me; I have zero lab experience). In poth the above plots, we can see that the means (represented by the thick black lines in the boxplot) are fairly close, indicating that the data is consistent.

## Correlation analysis (cor.test)

```r
# Strep vs. prev
cor.test(log(strep), sqrt(prev))

# Strep vs. fuso
cor.test(log(strep), fuso)

# Fuso vs. lepto
cor.test(log(strep), sqrt(lepto))
```

### Part D: Log2Fold analysis

Run the log2fold analysis just like in the pdf. Use the markdown to comment on what is happening

```r
a <- 10
b <- 100
fc <- b/a

fc

a <- 1:10
b <- 10:1
fc <- b/a
df <- data.frame(a, b, fc, a.greaterthan.b = a > b)
df

ggplot(df, aes(a, fc, colour = a.greaterthan.b), size = 8) +
      geom_point()

log2fc <- log2(fc)
ggplot(df, aes(a, log2fc, colour = a.greaterthan.b), size = 8) +
    geom_point()

ggplot(df, aes(fc, log2fc, colour = a.greaterthan.b), size = 8) +
    geom_point()
```
